Splitting the data into 3 and storing to JSON EXCEL CSV formats

```python
total_rows = len(df)
```

```python
# Calculate the number of rows for each split
split_1 = total_rows // 3
split_2 = (total_rows - split_1) // 2
```

```python
# Split the DataFrame into three parts
df1 = df.iloc[:split_1]
df2 = df.iloc[split_1:split_1 + split_2]
df3 = df.iloc[split_1 + split_2:]
```

```python
#saving df2 in json formate
df1.to_csv('df1_columns.csv')
```

```python
#saving df2 in json formate
df2.to_json('df2_columns.json', orient='columns')
```

```python
#saving df3 in json formate
df3.to_excel('df3_columns.xlsx', engine='openpyxl')
```

```python
#Droping Duplicate rows
df1.drop_duplicates(inplace=True)
df2.drop_duplicates(inplace=True)
df3.drop_duplicates(inplace=True)
```

**Type casting -  Standardizing Data Types**

date is stored in object formate that we need to convert into datetime format

```python
# Convert 'Date' column to datetime format, specifying dayfirst=True
df1['Date'] = pd.to_datetime(df1['Date'], dayfirst=True)
df2['Date'] = pd.to_datetime(df2['Date'], dayfirst=True)
df3['Date'] = pd.to_datetime(df3['Date'], dayfirst=True)
```

**Data Transformation**


IsHolyday Column is True and False only , We need to convert 

```python
from sklearn.preprocessing import LabelEncoder

# Initialize the label encoder
label_encoder = LabelEncoder()

# Assuming 'IsHoliday' column contains 'Yes' and 'No' or similar categorical values
df1['IsHoliday'] = label_encoder.fit_transform(df1['IsHoliday'])
df2['IsHoliday'] = label_encoder.fit_transform(df2['IsHoliday'])
df3['IsHoliday'] = label_encoder.fit_transform(df3['IsHoliday'])

# Check the result
print(df1.head())
print(df2.tail())
print(df3.tail())
```

Concatenating Data from Different Sources

```python
df_combined = pd.concat([df1, df2, df3], ignore_index=True)

# Final inspection of the combined data
print(df_combined.info())
print(df_combined.head())
```
